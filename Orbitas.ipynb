```python
import numpy as np
import matplotlib.pyplot as plt
```

#Ejemplo 1

```python
f = lambda x : np.sqrt(x)

n = 10

seeds = [0.5, 3.2, 1, 0, 6, 7, 4, 3.5, 4 , 11, 0.4]

orbits = []

for seed in seeds:
    orbit = [seed]
    for i in range(n):
        orbit.append(f(orbit[-1]))
    orbits.append(orbit)
orbits

for orbit in orbits:
    plt.plot(range(n + 1), orbit, marker='>')

plt.grid()
```

#Ejemplo 2

```python
f = lambda x: x ** 2

n = 10

seeds = [0.5, 0.2, 0.9, 0.5]

orbits = []

for seed in seeds:
    orbit = [seed]
    for i in range(n):
        orbit.append(f(orbit[-1]))
    orbits.append(orbit)
orbits

for orbit in orbits:
    plt.plot(range(n + 1), orbit, marker='o')

plt.grid()
```
