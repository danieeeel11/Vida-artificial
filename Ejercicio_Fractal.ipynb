## Ejercicio Fractales:

Construya un programa que realice lo siguiente:

1. Seleccionar un punto al azar al interno del triangulo ABC

2. Seleccionar aleatoriamente uno de los tres vertices (todos con la misma probabilidad)

3. Encontrar el punto medio entre el vertice y el punto seleccionado

4. Repetir pasos 2 y 3 100000 veces

```
import matplotlib.pyplot as plt
import numpy as np
```

```
def dibujar_triangle(p_ini, p_fin):

    ancho_base = p_fin[0] - p_ini[0]

    altura = (np.sqrt(3) / 2) * ancho_base

    mitad = (ancho_base / 2) + p_ini[0]

    # Crear el gráfico
    plt.plot( [p_ini[0], p_fin[0]] , [p_ini[1], p_fin[1]], color='red')
    plt.plot( [p_ini[0], mitad] , [p_ini[1], altura], color='red')
    plt.plot( [p_fin[0], mitad] , [p_fin[1], altura], color='red')

    # Etiquetar los vértices

    plt.text(p_ini[0], p_ini[1]-0.1, 'A', fontsize=12, ha='center')
    plt.text(p_fin[0], p_fin[1]-0.1, 'B', fontsize=12, ha='center')
    plt.text(mitad, altura + 0.05, 'C', fontsize=12, ha='center')

    # Mostrar el gráfico
    plt.show()

    return [ p_ini, p_fin, [mitad,altura] ]
```

```
# Función para seleccionar un vértice al azar con igual probabilidad
def vertice_random():
    return np.random.choice(len(vertices))
```

```
# Función para encontrar el punto medio entre dos puntos
def punto_medio(p1, p2):
    return (p1 + p2) / 2
```

```
vertices = dibujar_triangle([1,0], [4,0])
```

```
vertices = np.array(vertices)
```

```
# Número de iteraciones
n_iteraciones = 100000
# Almacenar los puntos medios en una lista
puntos_medios = []
```

```
# Selecciona un punto aleatorio dentro del triangulo
r1, r2 = np.random.rand(2)
punto = (1 - np.sqrt(r1)) * vertices[0] + np.sqrt(r1) * (1 - r2) * vertices[1] + np.sqrt(r1) * r2 * vertices[2]
```
